# Grover's Algorithm Tutorial

This notebook provides an interactive demonstration of Grover's quantum search algorithm.

## Setup

First, let's import the necessary libraries and set up our environment:

```python
import numpy as np
from grover_search_engine.grover import grovers_algorithm
from grover_search_engine.quantum_circuit import create_grover_circuit
import matplotlib.pyplot as plt
from ipywidgets import interact, IntSlider, Dropdown
```

## Interactive Grover's Algorithm Demonstration

Use the widgets below to adjust parameters and visualize the algorithm in action:

```python
# Define interactive widgets
qubit_widget = IntSlider(min=2, max=8, value=3, description='Number of qubits:')
marked_state_widget = IntSlider(min=0, max=7, value=1, description='Marked state (decimal):')
iterations_widget = IntSlider(min=1, max=5, value=2, description='Grover iterations:')

# Create interactive function
def run_grover(qubits, marked_state, iterations):
    """Run Grover's algorithm with given parameters and visualize results."""

    # Convert marked state to binary string
    marked_binary = format(marked_state, f'0{qubits}b')

    print(f"\nParameters:")
    print(f"- Number of qubits: {qubits}")
    print(f"- Marked state (binary): {marked_binary}")
    print(f"- Grover iterations: {iterations}")

    # Create and run quantum circuit
    circuit = create_grover_circuit(qubits, marked_state, iterations)
    result = grovers_algorithm(circuit)

    # Visualize the results
    plt.figure(figsize=(10, 6))

    # Plot probability distribution
    probabilities = np.abs(result)**2
    bars = plt.bar(range(2**qubits), probabilities)

    # Highlight the marked state
    bars[marked_state].set_color('red')

    plt.xlabel('State (decimal)')
    plt.ylabel('Probability')
    plt.title(f'Grover Algorithm Results ({qubits} qubits, {iterations} iterations)')
    plt.grid(True, alpha=0.3)

    # Add value labels
    for i, prob in enumerate(probabilities):
        plt.text(i, prob + 0.01, f'{prob:.2f}', ha='center', fontsize=8)

    plt.show()

    return result

# Create interactive interface
interact(run_grover,
         qubits=qubit_widget,
         marked_state=marked_state_widget,
         iterations=iterations_widget);
```

## Explanation of Grover's Algorithm

Grover's algorithm is a quantum algorithm that provides quadratic speedup for unstructured search problems. Here's how it works:

1. **Initialization**: Start with all qubits in superposition state (|0⟩ + |1⟩)/√2
2. **Oracle**: Mark the target state by applying a phase flip
3. **Diffusion**: Amplify the amplitude of the marked state using diffusion operator
4. **Repeat**: Apply oracle and diffusion multiple times to increase probability of measuring the marked state

The optimal number of iterations is approximately π/4 * √N, where N is the search space size.

## Mathematical Background

For a system with N states:
- Each Grover iteration increases the probability of measuring the marked state
- After ~π/4 * √N iterations, the probability approaches 1
- The algorithm achieves O(√N) time complexity vs classical O(N)

```python
# Example: Calculate optimal iterations for different qubit counts
def calculate_optimal_iterations(qubits):
    """Calculate theoretical optimal number of Grover iterations."""
    N = 2**qubits
    return int(np.round((np.pi/4) * np.sqrt(N)))

print("\nOptimal iterations for different qubit counts:")
for q in range(2, 9):
    print(f"{q} qubits: {calculate_optimal_iterations(q)} iterations (search space size: {2**q})")
```

## Advanced Visualization

Let's visualize how the probability distribution evolves with each Grover iteration:

```python
# Advanced visualization function
def visualize_grover_evolution(qubits, marked_state, max_iterations):
    """Visualize probability evolution across multiple iterations."""

    plt.figure(figsize=(12, 8))

    for iteration in range(1, max_iterations + 1):
        circuit = create_grover_circuit(qubits, marked_state, iteration)
        result = grovers_algorithm(circuit)
        probabilities = np.abs(result)**2

        plt.plot(range(2**qubits), probabilities, 'o-', label=f'Iteration {iteration}')

    plt.axvline(x=marked_state, color='red', linestyle='--', label='Marked state')
    plt.xlabel('State (decimal)')
    plt.ylabel('Probability')
    plt.title(f'Grover Algorithm Evolution ({qubits} qubits, marked state {marked_state})')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

# Interactive visualization
interact(visualize_grover_evolution,
         qubits=IntSlider(min=2, max=6, value=3, description='Qubits:'),
         marked_state=IntSlider(min=0, max=7, value=1, description='Marked state:'),
         max_iterations=IntSlider(min=1, max=5, value=4, description='Max iterations:'));
```
